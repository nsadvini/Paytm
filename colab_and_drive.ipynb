```python
import numpy as np
employee_rat['log_income']=np.log(employee_rat['MonthlyIncome'])
employee_rat[["log_income","MonthlyIncome"]]
```

```python
print(list(set(employee_rat['EnvironmentSatisfaction'])))
print(list(set(employee_rat['PerformanceRating'])))
```

```python
Env= {'Low' : 1, 'Medium' : 2, 'High' : 3, 'Very High':4}
perf={'Excellent':1, 'Outstanding':2}

employee_rat['EnvironmentSatisfaction']=employee_rat['EnvironmentSatisfaction'].map(Env)
employee_rat['PerformanceRating']=employee_rat['PerformanceRating'].map(perf)
```
